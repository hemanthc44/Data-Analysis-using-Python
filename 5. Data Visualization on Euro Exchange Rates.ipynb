# Introducing the dataset 

```python
import pandas as pd

ex_rate = pd.read_csv('Downloads/Datasets/euro-daily-hist_1999_2022.csv')
ex_rate.head(10)
```

```python
ex_rate.info()
```

```python
ex_rate.rename(columns = {'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
              inplace = True)
ex_rate['Time'] = pd.to_datetime(ex_rate['Time'])
ex_rate.sort_values('Time', inplace=True)
ex_rate.reset_index(drop=True, inplace=True)
ex_rate.head()
```

```python
euro_to_dollar = ex_rate[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

# Rolling mean

```python
import matplotlib.pyplot as plt

plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
```

```python

plt.figure(figsize=(18,12))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip ([2,3,4,5,6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

# Data Visualization 

## The 2008 Financial crisis

```python
financial_crisis = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2006
                   ) & (euro_to_dollar['Time'].dt.year <= 2009)]
financial_crisis_7_8 = euro_to_dollar.copy(
                   )[(euro_to_dollar.Time.dt.year >= 2007
                   ) & (euro_to_dollar.Time.dt.year <= 2008)]
```

```python
### Adding the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

### Adding the plot
fig,ax = plt.subplots(figsize=(8,3))
ax.plot(financial_crisis['Time'],
        financial_crisis['rolling_mean'],
        linewidth=1, color='#A6D785')

### Highlighting the 2007-2008 period
ax.plot(financial_crisis_7_8['Time'],
        financial_crisis_7_8['rolling_mean'],
        linewidth=3, color='#e23d28')

ax.set_xticklabels(['2006', "", '2007','', '2008','', '2009','', '2010'])

## Adding a title and a subtitle    
ax.text(13000, 1.67, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
        weight='bold')
ax.text(13000, 1.63, 'Euro-USD exchange rates between 2006 and 2010',
        size=12)


plt.show()
```
